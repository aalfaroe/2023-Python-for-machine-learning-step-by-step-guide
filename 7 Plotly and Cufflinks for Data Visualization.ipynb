# Plotly and Cufflinks for Data Visualization 

plotly.py is an interactive, open-source, and browser-based graphing library for Python ✨

plotly: https://github.com/plotly/plotly.py

cufflinks: https://github.com/santosjorge/cufflinks

```python
# !pip install plotly
# !pip install cufflinks
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as py
from plotly.offline import iplot
import cufflinks as cf
```

```python
py.offline.init_notebook_mode(connected=True)
cf.go_offline()

plt.rcParams['figure.figsize'] = [8,4]
plt.rcParams['figure.dpi'] = 300
```

```python
df = sns.load_dataset('iris')
df = df.drop('species', axis=1)
```

```python
df.iplot()
```

#### Scatter Plot

```python
df.iplot(x='sepal_length', y='sepal_width', mode='markers',
        size=15, title='Scatter Plot', yTitle='Sepal Width', xTitle='Sepal Length')
```

#### Bar Plot

https://www.kaggle.com/datasets/vinicius150987/titanic3

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.info()
```

```python
titanic.iplot(x='sex', y='survived', kind='bar', title='Titanic Dataset Analysis')
```

```python
titanic['sex'].value_counts()
```

```python
titanic[titanic['survived']==1]['sex'].value_counts()
```

```python
df.iplot(kind='bar', barmode='stack',
              title='IRIS Dataset Analysis')
```

```python
df.head(10).iplot(kind='barh', barmode='stack', bargap=0.1)
```

#### Box and Area Plot

```python
df.iplot(kind='box')
```

```python
df.iplot(kind='area', fill=True)
```

#### 3D Plot

```python
cf.colors.scales() 
```

```python
df.iplot(kind='surface', colorscale='rdylbu')
```

```python
cf.datagen.sinwave(10, 0.1).iplot(kind='surface')
```

#### Hist Plot, Bubble Plot and Heatmap

```python
df.iplot(kind='hist', bins=20, barmode='relative')
```

```python
bb = cf.datagen.bubble3d(5,4, mode='stocks')
```

```python
bb.iplot(kind='bubble3d', x='x', y='y', z='z', size='size')
```

```python
df.corr()
```

```python
df.corr().iplot(kind='heatmap', colorscale='spectral')
```
