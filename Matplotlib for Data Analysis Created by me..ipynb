```python
plt.figure(figsize=(10,3))
#plt.plot(movies['Rank'], movies['Runtime (Minutes)'])
plt.plot(movies['Rank'], movies['Rating'], color='blue', linestyle='--',
         linewidth=1.5, marker='o', markersize=8, markerfacecolor='y',
         markeredgecolor='blue', markeredgewidth=2)
plt.xlabel('Rank')
plt.ylabel('Rating')
plt.title('Rank vs Rating')
plt.show()
```

#Subplots

```python
#2x2 subplots another way

#plt.figure(figsize=(8,6))
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(8,6))

ax[0][0].plot(movies['Rank'], movies['Rating'], color='blue', linestyle='--',
         linewidth=1.5, marker='o', markersize=8, markerfacecolor='darkorange',
         markeredgecolor='blue', markeredgewidth=2)
ax[0][0].set_xlabel('Rank')
ax[0][0].set_ylabel('Rating')
ax[0][0].set_title('Rank vrs Rating')

#--------------------------------------------------

ax[0][1].plot(movies['Rank'], movies['Revenue (Millions)'], color='g', linestyle='--',
         linewidth=1.5, marker='o', markersize=8, markerfacecolor='orange',
         markeredgecolor='lightblue', markeredgewidth=2)
ax[0][1].set_xlabel('Rank')
ax[0][1].set_ylabel('Revenue (Millions)')
ax[0][1].set_title('Rank vrs Revenue (Millions)')

#---------------------------------------------------

ax[1][0].plot(movies['Rank'], movies['Metascore'], color='brown', linestyle='--',
         linewidth=1.5, marker='o', markersize=8, markerfacecolor='y',
         markeredgecolor='purple', markeredgewidth=2)
ax[1][0].set_xlabel('Rank')
ax[1][0].set_ylabel('Metascore)')
ax[1][0].set_title('Rank vrs Metascore')

#----------------------------------------------------


ax[1][1].plot(movies['Rank'], movies['Votes'], color='black', linestyle='--',
         linewidth=1.5, marker='o', markersize=8, markerfacecolor='w',
         markeredgecolor='m', markeredgewidth=2)
ax[1][1].set_xlabel('Rank')
ax[1][1].set_ylabel('Votes')
ax[1][1].set_title('Rank vrs Votes')         
          
#ax.tight_layout() #esto es para q los graficos esten a la misma medida
plt.tight_layout()
```

#Creating a Zoomed Sub-Figure of a Figure

```python
fig = plt.figure(figsize=(10,3))
ax1 = fig.add_axes([0,0, 1,1])
ax2 = fig.add_axes([0.1,0.1, 0.15,0.35])

ax1.plot(movies['Rank'], movies['Rating'], color='blue', linestyle='--',
         linewidth=1.5, marker='o', markersize=8, markerfacecolor='darkorange',
         markeredgecolor='blue', markeredgewidth=2)
ax1.set_xlabel('Rank')
ax1.set_ylabel('Rating')
ax1.set_title('Rank vrs Rating')

#---------------------------------------------------
#plot zoomed subplots

ax2.plot(movies['Rank'], movies['Rating'], color='blue', linestyle='--',
         linewidth=1.5, marker='o', markersize=8, markerfacecolor='darkorange',
         markeredgecolor='blue', markeredgewidth=2)

ax2.set_xlim(41,45)
ax2.set_ylim(1,9)
ax2.set_xlabel('Rank')
ax2.set_ylabel('Rating')
ax2.set_title('Rank vrs Rating (Zoomed)')

plt.show()
```

#xlim and ylim, legend, grid, xticks, yticks

```python
plt.figure(figsize=(10,3))
plt.plot(movies['Rank'], movies['Rating'], linestyle='--', color='b', 
         label='Rating')

plt.xlim(0,45) #limita el rango del eje x de cero a 30
plt.ylim(4,10)  #limita el rango del eje y de cuatro a 10
plt.xlabel('Rank')
plt.ylabel('Rating')
plt.title('Rank vrs Rating')

plt.legend()
plt.grid()#hace como el cuadriculado del plot

plt.xticks(np.arange(0,45, 4)) #en el eje x se ponen los numeros del 0 a 30 con brincos de 2 en 2
plt.yticks(np.arange(0,10, 1))
plt.show()
```
